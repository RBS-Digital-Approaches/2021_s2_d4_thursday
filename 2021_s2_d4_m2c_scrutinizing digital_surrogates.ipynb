# Scrutinizing digital surrogates
For the first part of this session, I'm asking you to examine (portions of) a few different digital surrogates with the same kind of focused attention that analytical bibliographers bring to the examination of material texts. 

There are a few broad questions I'm hoping to set up by starting from the examination of these images:

* How did these images come to look the way they do?
* What of what we're seeing when we look at these surrogates do we take to be:
  - Properties of the book that we would be trying to examine if we had access to the item, itself;
  - Artifacts produced by photography;
  - Artifacts produced by computational processing of the image.
* How do we begin to distinguish among those things?
* How far can an informed alertness about the properties of digital surrogates allow us to use them for certain kinds of bibliographical and book historical work?

## 1 - James Thomson, *The Tragedy of Sophonisba* (Oxford English Faculty Library XL78.1[SOP])
We'll end up looking at surrogates of a few different copies of this play (that I have collated pretty extensively, but have so far managed to avoid reading).  

I have a few questions I'd like you to consider where this particular copy is concerned.

1. First, please have a look at the version that you'll find at [Google Books](https://books.google.com/books?id=81QJAAAAQAAJ&printsec=titlepage#v=onepage&q&f=false). I'd like you to look, in particular, at pp. 49-56 and try to arrive at an hypothesis about what we're seeing there. You might find it useful to move back and forth between Google Books' one-up and two-up views. Or you might not. I've placed some pages from Philip Gaskell's *A New Introduction to Bibliography* in the shared folder that Rare Book School created for the class that you may find helpful to consult.
2. While still at Google Books, please also have a look at the version of this book that you can download from the site. Click on the gear icon at the upper right, and select "Download PDF" (if you right-click on "Download PDF" you can open the PDF in a new window or tab, rather than downloading the file to your hard drive.)
  Take some time scroll through and compare the downloadable file to what you're seeing in your browser. There's the obvious fact that the downloadable file is black and white, while the in-browser view is in color. But keep looking around the book to see what ideas you might begin to form about the relationships between these two versions. Both of these are derivatives of an original set of images that Google photographed at the Bodleian (or, potentially, *two* original sets of images, since Google typically photographed each page twice, as I understand). How do these two versions, both available from Google, compare?
3. Keep the versions from Google handy, but also have a look at the [PDF that Oxford makes available](http://dbooks.bodleian.ox.ac.uk/books/PDFs/400247933.pdf) (findable through its SOLO catalogue). Spend some time comparing this black and white PDF to the downloadable PDF from Google Books. You don't have to be too minute in your comparison, but do take the time to really look at the two, side by side, and a page at a time. (Some pages will reward attention more than others.) What kinds of differences do you see between the two? Where do you see those differences?
4. Finally, I'd like you to look specifically at pages 39 and 42 in all three versions (the in-browser color version and the two black and white versions). What can you see in the color version that you can't see in the black and white versions? What might we the color presentation tell us about this particular item that we wouldn't learn from the black and white versions?
